#Transformer-Based Sentiment Classification using DistilBERT

## Overview
This project uses a transformer-based language model to classify text sentiment. The goal is to explore how modern NLP models such as BERT can capture contextual meaning in text more effectively than traditional machine learning approaches.

## Research Question
How effectively can transformer-based models classify sentiment in text compared to traditional machine learning approaches in terms of accuracy and contextual understanding?

```python
!pip install transformers datasets torch scikit-learn pandas

import pandas as pd
from datasets import load_dataset
from transformers import pipeline
from sklearn.metrics import accuracy_score
```

```python
#Load Dataset
dataset = load_dataset("imdb")

test_data = dataset["test"].shuffle(seed=42).select(range(200))

texts = test_data["text"]
labels = test_data["label"]

df = pd.DataFrame({"text": texts, "true_label": labels})
df.head()
```

## Dataset
The dataset used is the IMDB movie review dataset from the Hugging Face datasets library. Reviews are labeled as either positive or negative sentiment.

For this project, a subset of the test set is used to evaluate a transformer-based sentiment classification model.

```python
#Load BERT-style sentiment pipeline
classifier = pipeline("sentiment-analysis")
```

```python
#Generate predictions
predictions = classifier(list(df["text"][:50]), truncation=True)
predictions[:5]
```

```python
#Convert labels
predicted_labels = [1 if pred["label"] == "POSITIVE" else 0 for pred in predictions]
true_labels = list(df["true_label"][:50])

accuracy = accuracy_score(true_labels, predicted_labels)
print("Accuracy:", accuracy)
```

```python
#Show sample predictions
sample_df = pd.DataFrame({
    "text": list(df["text"][:5]),
    "true_label": true_labels[:5],
    "predicted_label": predicted_labels[:5],
    "confidence": [round(pred["score"], 4) for pred in predictions[:5]]
})

sample_df
```

#Results
The transformer-based model achieved high accuracy (0.94) on the evaluation subset of the IMDB dataset, demonstrating strong performance in sentiment classification tasks.

Sample predictions indicate that the model is able to correctly classify both positive and negative reviews with high confidence. The results demonstrate a clear improvement over traditional machine learning approaches used in earlier projects.

#Interpretation
The results show that transformer-based models are highly effective for sentiment classification tasks. By leveraging contextual language representations, the model can capture nuances in text that simpler models often miss.

Compared to traditional approaches such as TF-IDF with Logistic Regression, the transformer model achieves higher accuracy and provides more reliable predictions, particularly for complex or context-dependent sentences.

#Limitations
- Only a small subset of the dataset was used for evaluation
- No fine-tuning of the transformer model was performed
- The model may struggle with very long or highly ambiguous text
- Computational cost is higher compared to traditional machine learning models
